# data types

All real world objects and ideas are transformed into computational
format, into *types of data*. There are many data types in Python, but a
few of them will be important for us in this workshop.

We use the `type()` function to check data types. A function is a way of doing something to data in Python. We will talk more about functions in upcoming lessons. 

The most common data types include: 

-   `integer` 
-   `string`
-   `boolean` 
-   `list` 

Other types you might hear about: floats, tuples, dicts, etc. You can
read more about these [on
W3Schools](https://www.w3schools.com/python/python_datatypes.asp).

## int, str, bool, list

```python
type(1)
```

The first type is `int` which refers to integer, a whole number. (As opposed to a number with a decimal, which has its own data type)

```python
type('hello')
```

The next type is ~str~ which refers to string, or a sequence of characters (including letters, numbers, punctuation, and symbols) enclosed by quotes. To a human, a string may look just like language or words, but to a computer which doesn't understand human language, it is simply a sequence of characters, or a *string* of characters, between two quotes. 

```python
type('True')
```

The `bool` data type stands for boolean, a `True` or `False` value. Why do booleans get their own data type? Because truth and falsity are the foundation of more complex conditional statements, where certain actions will be executed depending on if a value is `True` or `False`.

```python
type(['donut', 'smoothie', 'coffee', 'banana'])
```

Finally, the `list` data type describes a collection of items, such as  a list of strings, like in this example. Lists can also contain integers, other lists, and items from more data types. Lists will be the most important data type for our work with text, as text-based data is often organized as strings within lists. More on this soon.

## data as constraints

Data types create a particular kind of *constraint* on data. First, all information that you want to work with in Python must fit into one of these data types. Transforming real world information into computer-readable data inevitably reduces or simplifies the original. It might seem too obvious to point out, but computers (not even with large-language models) do not understand human language the way humans understand it. Everything that humans consider a "word" becomes a `string` to a computer. It becomes a sequence of characters wrapped by quotation marks. A `string` like "ocean" has no semantic meaning to a computer, though a large enough computer (like a large-language model) can process enough strings to get a sense of where "ocean" should fall in a sentence, and what words should surround it, perhaps "deep" or "blue". But this kind of understanding is a statistical inference, and doesn't mean that the computer associates depth or color with the ocean, not to mention the various sensations that such words can evoke in humans. 

That being said, these kind of constraints can also be productive ones. Each data type comes with a number of ~methods~ that can be used on that particular type. You can do things with integers that you cannot do with strings, for example. You can add numbers together, but not strings. 

While you're still a beginner, the differences in data types will likely create headaches as you learn by errors (many, many errors) what methods can and cannot work with certain types of data. Data types will feel like a nuisance, preventing you from doing what you want to your data. But as you advance, you'll learn to leverage the different types to manipulate data in amazing ways.
