# exploring bs4

## what is bs4?

bs4 is short for `BeautifulSoup4`, a python package for parsing HTML data. bs4's power comes from using python syntax to access and manipulate HTML elements. This means that it uses the python language and its syntax to get information from pages written in the web's main computer lanugage, HTML.


```python
# import the following libraries to use bs4

import requests # to make https requests
from bs4 import BeautifulSoup # our web scraping library
import lxml # a parser for working with html data
```

```python
# save the data from the website as a "soup" object

site = requests.get('https://translegislation.com/bills/2023/passed')
html_code = site.content
soup = BeautifulSoup(html_code, 'lxml')
```

Using dot syntax, we can access the element contained within the `soup` object

```python
# get title

soup.title
```

## the `soup` object

This word "object" in Python is something you'll hear often. It means a collection of data and functions that can work on that data. You can think of it as a way of representing real world objects (like this web page) in a way that is organized and accessible, so you can work with that information, with Python.

Let's take an initial look into what this beautiful soup object can do. It takes the HTML source, the specific HTML elelements or "tags," and makes it possible for us to access those tags using python syntax.

```python
soup.h2
```

```python
soup.a
```

There are multiple ways for accessing the soup object in Python. The Python syntax offers us a few options. For example, We might access the title attribute using dot syntax, like soup.title, or we might use the find() function, to access the title, like soup.find('title').

## getting text

Let's go a little deeper than the element. We can access just the text (getting rid of tags like `<p>` or `<h3>` from elements.

```python
soup.title.text
```

Combine this with variables, and we can save just the text.

```python
bill_title = soup.h3.text
```

```python
bill_title
```

## getting attributes

In addition to text, we can also get the attributes. Attributes are additional data that is contained within the HTML tag, like href, which stands for hyperlink reference. This is the link's URL. To access the attributes like href, we use the syntax: `tag['attr']`.

```python
soup.h3['class']
```

```python
link_location = soup.a['href']
```

```python
link_location
```

## `find()`

```python
soup.find('p')
```

You can also save what you find as a new variable. This is really useful for later, when we will write code to work with soup elements.

```python
paragraph = soup.find('p')
```

```python
paragraph
```

Making variables is useful for layering other operations on top, like getting `text`.

```python
paragraph.text
```

```python
link = soup.find('a')
print(link)
```

You can also use `find()` to search by specific attribute. Just include the `class_=xxx` in your `find()` call.

```python
soup.find('div', class_='css-wd7aku')
```

Combine this with the `text` property to get just the text

```python
soup.find('div', class_='css-wd7aku').text
```

## `find_all()`

Want to print out all tags of a specific element? Then we use `find_all()`

```python
soup.find_all('h3')
```

```python
headers = soup.find_all('h3')
```

```python
# remember list slicing? Here we use list slicing to print out only the first three elements

headers[:3]
```
