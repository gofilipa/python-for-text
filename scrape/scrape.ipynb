## what is bs4?
bs4 is short for `Beautiful Soup 4`, a python package for parsing HTML data. bs4's power comes from using python syntax to access and manipulate HTML elements. In other words, this means that it uses the python language and its syntax to get information from pages written in the web's main computer lanugage, HTML.

```python
import requests
from bs4 import BeautifulSoup
import lxml
import csv
```

```python
site = requests.get('https://translegislation.com/bills/2023/passed')
html_code = site.content
soup = BeautifulSoup(html_code, 'lxml')
```

```python
# for each item in div.css-4rck61
# get h3.text
# get category text class = css-bu60l4
# get h2.text
# get a class "chakra-link"
```

```python
# get bill title

soup.h3.text
```

```python
# get bill caption
soup.find('div', class_ ='css-4rck61').h2.text
```

```python
# get bill category

soup.find('div', class_='css-4rck61').span.text
```

```python
# get bill description (if any)

soup.find('div', class_ ='css-4rck61').p.text
```

```python
# get link extension

soup.find('div', class_ ='css-4rck61').a['href']
```

```python
bill_cards = soup.find_all('div', class_ ='css-4rck61')
```

```python
for div in bill_cards:
    print(div.h3.text)
    print(div.h2.text)
    print(div.span.text)
    print(div.p.text)
    print(div.a['href']) # add https://translegislation.com/bills/2023/US
```

```python
# next steps: 
# assign a variable for each item. 

for item in bill_cards:
    title = item.h3.text
    caption = item.h2.text
    category = item.find('span').text
    description = item.p.text
    link = 'https://translegislation.com/bills/2023/passed' + item.a['href']
    print(title, caption, category, description, link)
```

## getting the link to state bill

```python
## make a list of URLs to go to
## for each URL, make a soup.
## for each soup, get the link to the state bill
## add the link to the spreadsheet

# I want it to loop through each card item on this page, then
# when it gets to the link part, I want it to go into that page
# and grab the data from there. 
# fuck I just need to append a column, don't I? 
```

```python
for item in bill_cards:
    extension = 'https://translegislation.com/' + item.a['href']
    print(extension)
```

```python
urls = []
for item in bill_cards:
    extension = 'https://translegislation.com/' + item.a['href']
    urls.append(extension)
```

```python
urls[:10]
```

```python
# this may take several seconds

soups = []
for item in urls:
    site = requests.get(item)
    html_code = site.content
    soup = BeautifulSoup(html_code, 'lxml')
    soups.append(soup)
```

```python
legiscan_links = []
for item in soups:
    # get the url for state url
    anchor_tag = item.find('a', class_='chakra-link css-oga2ct')
    link = anchor_tag['href']
    legiscan_links.append(link)
```
