# analyzing: generating questions
It's a good idea to have a sense of what you're looking for before you jump into analysis. Though all text analysis (and any kind of data analysis more generally) can be a recursive process of formulating questions, doing some kind of manipulation, shifting your focus, and then running the analysis again.

Text analysis is a speculative and creative endeavor, because it requires the human running the program to imagine what kinds of patterns they might find in the data. The process of analysis itself will bring you to re-frame your perspective and re-write your questions. Though it feels like a step backward, this recursive process means that you're making progress toward the goal. 

## defining gender
So, ask yourself before you begin analysis: what are some of the most important things you want to know about the data? For my dataset, the important points are the following:
- How each bill is defining gender.
- The relationship between gender, sex, and sexuality.
- The common conflation of gender as a cultural and expressive phenomenon and sex as a biological one.

Informing these questions is an understanding of gender drawn from a linked vocabulary of LGBTQ+ terms, the "[Homosaurus Vocabulary](https://homosaurus.org/)." This vocabulary is designed to improve the description and discoverabiliy of LGBTQ+ information by offering definitions of various LGBTQ+ related terms. It defines gender as a cultural phenomenon, "A socially constructed system that gives meaning to masculinity and femininity," which is "developed through the interaction of social roles and expectations, one’s response to those expectations, one’s physiology, and one’s internal sense of self" ("Gender," Homosaurus). Sex, by contrast, is "The category assigned to an individual to describe their body and its role in sexual reproduction, usually based on external anatomy, but also including chromosomes, gene expression, hormones, and internal and external reproductive/sexual organs" ("Sex," Homosaurus).

## a rules-based approach
As a brief refresher of the previous section: after `spaCy` runs our text through the pipes, it creates a `Doc` object with annotations. These annotations contain information about each word in the document, its part of speech or entity label, for example, which we can leverage to write code.

Writing code that enables us to access text by their annotation is called a "rules-based" approach to text analysis. We create a set of "rules," or instructions, for the code to use to find text. For example, we can write code that can match based on whether a particular word or series of words appear in a text. 

In what follows, I'm doing to demonstrate three ways of writing rules to work with text in `spaCy`. Each of the three methods has their benefits and drawbacks. The ones that are easier to implement often don't come with as much potential for customization (which we will see with the `PhraseMatcher`, that can only find text which matches exactly what you are looking for). Other more complex processes, like the Token `Matcher`, can find the same word in a variety of forms, surrounded by different kinds of punctuation or its position in a sentence, for example.
