## remove some ents
- People's names: PERSON
- numbers: ORDINAL, CARDINAL, DATE, TIME

```python
def remove_ents(text):
    nents = []
    take_out = ['PERSON', 'ORDINAL', 'CARDINAL', 'DATE', 'TIME']
    for item in text.ents:
        if item.label_ not in take_out:
            nents.append(item)
    return nents
```

```python
with open('matcher_defs.txt', 'w') as f:
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  # Get string representation
        span = doc[start:end]  # The matched span
        f.write(str(span.text))
        f.write(str(doc[start].sent))
        f.write(str('\n'))
        f.write(str('\n'))
```

## moving on from spacy

After doing this, the type of object is NOT a spacy object, which means that you lose things like token attributes, named entity labels, sentence demarcations, parts of speech, etc. All of the "annotations" (so to speak) that spacy gives to its data when you process it with the pipes. 

By turning the Doc object into a regular string object (from which it came!), you're flattening the data to just characters within a sequence. But this also has its benefits. Now you can transport that data to other libraries for other kinds of analysis. For example, you can paste the text into Voyant. 

```python
for token in doc[:10]:
    print(token, token.like_num)
```

```python
def token_filter(token):
    return not (token.is_punct | token.is_space | token.is_stop | 
                token.like_url | token.like_num | token.like_email)
```

```python
plain_text = [] 
for token in doc:
    if token.is_alpha:
        if token_filter(token):
            lemma = token.lemma_.lower()
            plain_text.append(lemma)
```
